1. Escreva um aplicativo que exibe uma caixa, uma oval, uma seta e um
losango utilizando asteriscos (*).

```python
def exibir_formas():
    """Exibe uma caixa, uma oval, uma seta e um losango utilizando asteriscos."""
    
    def desenhar_caixa(altura=4, largura=7):
        for i in range(altura):
            if i == 0 or i == altura - 1:
                print('*' * largura)
            else:
                print('*' + ' ' * (largura - 2) + '*')
    
    def desenhar_oval(altura=5, largura=7):
        for i in range(altura):
            if i == 0 or i == altura - 1:
                print(' ' * ((largura - 3) // 2) + '***')
            elif i == 1 or i == altura - 2:
                print(' *' + ' ' * (largura - 4) + '* ')
            else:
                print('*' + ' ' * (largura - 2) + '*')
    
    def desenhar_seta(altura=6, largura=7):
        metade = largura // 2
        for i in range(altura):
            if i < 3:  # Cabeça da seta
                print(' ' * (metade - i) + '*' * (2 * i + 1))
            else:  # Haste da seta
                print(' ' * metade + '*')
    
    def desenhar_losango(altura=7, largura=7):
        metade = altura // 2
        for i in range(altura):
            if i <= metade:
                espacos = metade - i
                asteriscos = 2 * i + 1
            else:
                espacos = i - metade
                asteriscos = 2 * (altura - i) - 1
            print(' ' * espacos + '*' * asteriscos)
    
    print("Caixa:\n")
    desenhar_caixa()
    print("\nOval:\n")
    desenhar_oval()
    print("\nSeta:\n")
    desenhar_seta()
    print("\nLosango:\n")
    desenhar_losango()

# Para exibir as formas, execute:
exibir_formas()
```

2. Escreva um aplicativo que insere um número consistindo em cinco dígitos do usuário, separa o número em seus dígitos individuais e imprime os dígitos separados uns dos outros por três espaços cada. Por exemplo, se o usuário digitar o número 42339, o programa deve imprimir: 4 2 3 3 9.

```python
def separar_digitos():
    """
    Solicita um número de 5 dígitos do usuário e exibe cada dígito separado por três espaços.
    """
    while True:
        try:
            # Solicita a entrada do usuário
            numero = input("Digite um número de 5 dígitos: ")
            
            # Verifica se o número tem exatamente 5 dígitos e contém apenas números
            if not numero.isdigit() or len(numero) != 5:
                print("Por favor, insira exatamente 5 dígitos numéricos.")
                continue
                
            # Converte para inteiro para reminar zeros à esquerda
            numero = int(numero)
            
            # Separa os dígitos
            digitos = [d for d in str(numero).zfill(5)]
            
            # Junta os dígitos com três espaços entre eles
            resultado = "   ".join(digitos)
            
            # Exibe o resultado
            print(f"\nNúmero separado: {resultado}")
            break
            
        except ValueError:
            print("Entrada inválida. Por favor, tente novamente.")

# Para executar a função, descomente a linha abaixo:
separar_digitos()
```

3. Faça um programa que calcule a distância entre dois pontos no plano cartesiano
> d = sqrt((x2-x1)^2 + (y2-y1)^2)

```python
import math

def calcular_distancia():
    """
    Calcula a distância entre dois pontos no plano cartesiano.
    O usuário deve fornecer as coordenadas (x, y) para ambos os pontos.
    """
    print("Cálculo da distância entre dois pontos no plano cartesiano")
    print("----------------------------------------------------")
    
    try:
        # Solicita as coordenadas do primeiro ponto
        x1 = float(input("Digite a coordenada x do primeiro ponto: "))
        y1 = float(input("Digite a coordenada y do primeiro ponto: "))
        
        # Solicita as coordenadas do segundo ponto
        x2 = float(input("Digite a coordenada x do segundo ponto: "))
        y2 = float(input("Digite a coordenada y do segundo ponto: "))
        
        # Calcula a distância usando a fórmula da distância euclidiana
        distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        
        # Exibe o resultado formatado
        print(f"\nA distância entre os pontos ({x1}, {y1}) e ({x2}, {y2}) é: {distancia:.2f}")
        
    except ValueError:
        print("Erro: Por favor, insira valores numéricos válidos.")

# Para executar a função, descomente a linha abaixo:
calcular_distancia()
```

4. Crie um programa que peça ao usuário para digitar uma frase e conte
quantas vogais (a, e, i, o, u) aparecem na frase. Considere que a contagem
deve ser case-insensitive (ou seja, não diferencie maiúsculas de
minúsculas).

```python
def contar_vogais():
    """
    Conta o número de vogais em uma frase fornecida pelo usuário.
    A contagem é case-insensitive (não diferencia maiúsculas de minúsculas).
    """
    print("Contador de Vogais")
    print("-----------------")
    
    # Solicita a frase ao usuário
    frase = input("Digite uma frase: ")
    
    # Converte para minúsculas para fazer a contagem case-insensitive
    frase = frase.lower()
    
    # Define as vogais
    vogais = 'aeiouáéíóúâêîôûãõ'
    
    # Conta cada vogal
    contagem = {vogal: frase.count(vogal) for vogal in vogais if vogal in frase}
    
    # Exibe o resultado
    print("\n--- Resultado ---")
    print(f"Frase: {frase}")
    print(f"Total de vogais: {sum(contagem.values())}")
    
    print("\nDetalhamento por vogal:")
    for vogal, quantidade in sorted(contagem.items()):
        print(f"'{vogal}': {quantidade}")

# Para executar a função, descomente a linha abaixo:
contar_vogais()
```

5. Uma empresa que quer enviar dados pela internet pediu-lhe queescrevesse um programa que criptografe dados a fim de que eles possam
ser transmitidos com maior segurança. Todos os dados são transmitidos como inteiros de quatro dígitos. Seu aplicativo deve ler um inteiro de quatro dígitos inserido pelo usuário e criptografá-lo da seguinte maneira: substitua cada dígito pelo resultado da adição de 7 ao dígito, obtendo o restante depois da divisão do novo valor por 10. Troque então o primeiro dígito pelo terceiro e o segundo dígito pelo quarto. Então, imprima o inteiro criptografado. Escreva um aplicativo separado que receba a entrada de um inteiro de quatro dígitos criptografado e o descriptografe (revertendo o esquema de criptografia) para formar o número original.

```python
def criptografar_numero():
    """
    Criptografa um número de 4 dígitos seguindo o esquema:
    1. Adiciona 7 a cada dígito e obtém o resto da divisão por 10
    2. Troca o 1º dígito com o 3º e o 2º com o 4º
    """
    while True:
        try:
            numero = input("Digite um número de 4 dígitos para criptografar: ")
            if not numero.isdigit() or len(numero) != 4:
                print("Por favor, insira exatamente 4 dígitos numéricos.")
                continue
                
            # Converte cada dígito para inteiro
            digitos = [int(d) for d in numero]
            
            # Passo 1: Adiciona 7 e obtém o resto da divisão por 10
            cripto = [(d + 7) % 10 for d in digitos]
            
            # Passo 2: Troca as posições
            cripto[0], cripto[2] = cripto[2], cripto[0]
            cripto[1], cripto[3] = cripto[3], cripto[1]
            
            # Junta os dígitos
            numero_cripto = ''.join(map(str, cripto))
            
            print(f"\nNúmero criptografado: {numero_cripto}")
            return numero_cripto
            
        except ValueError:
            print("Entrada inválida. Tente novamente.")

def descriptografar_numero():
    """
    Descriptografa um número de 4 dígitos seguindo o esquema inverso:
    1. Desfaz a troca de posições
    2. Subtrai 7 de cada dígito (ou adiciona 3, que é equivalente no módulo 10)
    """
    while True:
        try:
            numero = input("Digite um número criptografado de 4 dígitos para descriptografar: ")
            if not numero.isdigit() or len(numero) != 4:
                print("Por favor, insira exatamente 4 dígitos numéricos.")
                continue
                
            # Converte cada dígito para inteiro
            digitos = [int(d) for d in numero]
            
            # Passo 1: Desfaz a troca de posições
            digitos[0], digitos[2] = digitos[2], digitos[0]
            digitos[1], digitos[3] = digitos[3], digitos[1]
            
            # Passo 2: Subtrai 7 (equivalente a somar 3 no módulo 10)
            original = [(d + 3) % 10 for d in digitos]
            
            # Junta os dígitos
            numero_original = ''.join(map(str, original))
            
            print(f"\nNúmero original: {numero_original}")
            return numero_original
            
        except ValueError:
            print("Entrada inválida. Tente novamente.")

def menu_criptografia():
    """
    Menu interativo para criptografar ou descriptografar números.
    """
    while True:
        print("\n--- Menu de Criptografia ---")
        print("1. Criptografar número")
        print("2. Descriptografar número")
        print("3. Sair")
        
        escolha = input("\nEscolha uma opção (1-3): ")
        
        if escolha == '1':
            criptografar_numero()
        elif escolha == '2':
            descriptografar_numero()
        elif escolha == '3':
            print("Encerrando o programa...")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Para executar o menu, descomente a linha abaixo:
menu_criptografia()
```
